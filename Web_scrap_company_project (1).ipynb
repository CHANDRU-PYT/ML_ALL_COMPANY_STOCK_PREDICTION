```python
import requests
from bs4 import BeautifulSoup
import csv
import time


def extract_all_pages():

    base_url = "https://www.screener.in/screens/357649/all-listed-companies/?page={}"

    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
    }

    session = requests.Session()
    session.headers.update(headers)

    with open("All_Companyreport.csv", "w", newline="", encoding="utf-8") as csv_file:
        csv_writer = csv.writer(csv_file)

        page = 1
        max_pages = 212   

        while page <= max_pages:
            try:
                print(f"Scraping Page {page}...")

                url = base_url.format(page)
                response = session.get(url, timeout=15)

                if response.status_code != 200:
                    print(f"Failed on page {page}")
                    page += 1
                    continue

                soup = BeautifulSoup(response.content, "lxml")

                table = soup.find("table")

                if table is None:
                    print(f"No table found on page {page}")
                    page += 1
                    continue

                if page == 1:
                    thead = table.find("thead")
                    if thead:
                        headers_row = [th.text.strip() for th in thead.find_all("th")]
                        csv_writer.writerow(headers_row)

                
                tbody = table.find("tbody")
                if tbody:
                    rows = tbody.find_all("tr")
                    for tr in rows:
                        row = [td.text.strip() for td in tr.find_all("td")]
                        csv_writer.writerow(row)

                time.sleep(5)  
                page += 1

            except requests.exceptions.RequestException as e:
                print(f"Connection error on page {page}: {e}")
                page += 1
                time.sleep(5)
                continue

            except Exception as e:
                print(f"Unexpected error on page {page}: {e}")
                page += 1
                continue

    print("Scraping Completed Successfully ✅")



extract_all_pages()
```
